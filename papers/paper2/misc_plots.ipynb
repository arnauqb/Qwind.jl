```julia
using Qwind, PyPlot
plt.style.use("science")
mnras_width = 252/72.27
```

```julia
model = Model("./config.yaml");
```

```julia
fig, ax = plt.subplots(figsize=(mnras_width, 0.8 * mnras_width))
r_range = model.rad.disk_grid
temps = disk_temperature.(Ref(model.bh), r_range)
line1, = ax.plot(r_range, model.rad.fuv_grid, linewidth=1.25)
ax2 = ax.twinx()
line2, = ax2.plot(r_range, temps, color= "C1", linewidth=1.25)
ax.set_xscale("log")
ax.legend([line1, line2], ["UV fraction", "Temperature"], loc="lower left", borderpad=1)
ax.set_xlabel(L"R [ $R_g$ ]")
ax.set_ylabel("UV fraction")
ax2.set_ylabel("Temperature [K]", rotation = -90, labelpad=15)
fig.savefig("./plots/uv_fractions.pdf")
```
